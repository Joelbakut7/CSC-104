```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\user\Downloads\all_recs.csv",
    encoding="unicode_escape"
)

df.head()
```

```python
df.tail(12)
```

```python
year_order = df["Year"].value_counts().index

plt.figure(figsize=(20,8))
plt.title("Attack Frequency by Year")

sb.countplot(
    data=df,
    y="Year",
    color=sb.color_palette()[2],
    order=year_order
)

plt.xticks(rotation=45)
plt.show()
```

```python
type_order = df["Type"].value_counts().index

plt.figure(figsize=(20,8))
plt.title("Attack Frequency by Year")

sb.countplot(
    data=df,
    y="Type",
    color=sb.color_palette()[2],
    order=type_order
)

plt.xticks(rotation=45)
plt.show()
```

```python
location_order = df["Location"].value_counts().index

plt.figure(figsize=(24,27))
plt.title("Attack Frequency by Location")

sb.countplot(
    data=df,
    y="Location",
    color=sb.color_palette()[2],
    order=location_order
)

plt.xticks(rotation=45)
plt.show()
```

```python
motive_order = df["Motive"].value_counts().index

plt.figure(figsize=(20,8))
plt.title("Attack Frequency by Motive")

sb.countplot(
    data=df,
    y="Motive",
    color=sb.color_palette()[2],
    order=motive_order
)

plt.xticks(rotation=45)
plt.show()
```
